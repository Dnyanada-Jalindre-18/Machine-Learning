# Handwritten Digit Recognition
* The MNIST dataset is a well-known dataset consisting of 28x28 grayscale images.
* We'll be using machine learning classification on the MNIST database of handwritten digits.
* We'll use following methods

        * K nearest neighbor
        * Logistic Regression
        * Linear SVC

## Loading Libraries and dataset

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv("HDR_train.csv")
dataset.head()
```

## Locating Dependent and Independent variables

```python
X = dataset.iloc[:,1:]      # 783 columns excluding label 
y = dataset.iloc[:,0]       # label column
```

```python
print("X-Shape : ",X.shape)
print("y-shape : ",y.shape)
```

```python
plt.figure(figsize=(4,4))
plt.imshow(X.iloc[13].values.reshape(28,28),cmap="gray",interpolation='none')
print(y[13])
```

```python
digits = X.iloc[1000:1024].values
plt.figure(figsize = (15,4))
for i in range(24):
    plt.subplot(3,8,i+1)
    plt.imshow(digits[i].reshape((28,28)), cmap = plt.cm.binary)
    plt.axis("off")
plt.show()
```

## Locating Training and Testing variables


```python
X_train = X[:10000]
y_train = y[:10000]

X_test = X[10000:11000]
actual = y[10000:11000]
```

# K nearest neighbor

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors= 5)

knn_model.fit(X_train, y_train)

knn_predicted = knn_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_knn = accuracy_score(actual, knn_predicted)
accuracy_knn
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg_model = LogisticRegression()

log_reg_model.fit(X_train, y_train)

log_reg_predicted = log_reg_model.predict(X_test)
```

```python
accuracy_log_reg = accuracy_score(actual, log_reg_predicted)
accuracy_log_reg
```

# Linear Support vector Classification

```python
from sklearn.svm import LinearSVC
```

```python
SVC = LinearSVC()

SVC.fit(X_train, y_train)

SVC_predicted = SVC.predict(X_test)
```

```python
accuracy_svc = accuracy_score(actual, SVC_predicted)
accuracy_svc
```
